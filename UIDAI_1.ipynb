```python
bio = bio.sort_values("date").reset_index(drop=True)
demo = demo.sort_values("date").reset_index(drop=True)
enrol = enrol.sort_values("date").reset_index(drop=True)
```

```python
# Total updates per row
bio["total_bio_updates"] = bio["bio_age_5_17"] + bio["bio_age_17_"]
demo["total_demo_updates"] = demo["demo_age_5_17"] + demo["demo_age_17_"]

# Total enrolments per row
enrol["total_enrolments"] = (
    enrol["age_0_5"] +
    enrol["age_5_17"] +
    enrol["age_18_greater"]
)
```

```python
# Create year-month column
for df in [bio, demo, enrol]:
    df["year_month"] = df["date"].dt.to_period("M")
```

```python
bio_sm = (
    bio.groupby(["state", "year_month"])["total_bio_updates"]
    .sum()
    .reset_index()
)

demo_sm = (
    demo.groupby(["state", "year_month"])["total_demo_updates"]
    .sum()
    .reset_index()
)

enrol_sm = (
    enrol.groupby(["state", "year_month"])["total_enrolments"]
    .sum()
    .reset_index()
)
```

```python
state_month = enrol_sm.merge(
    demo_sm, on=["state", "year_month"], how="left"
).merge(
    bio_sm, on=["state", "year_month"], how="left"
)

# Replace missing values with 0
state_month[[
    "total_demo_updates",
    "total_bio_updates"
]] = state_month[[
    "total_demo_updates",
    "total_bio_updates"
]].fillna(0)

# Total updates
state_month["total_updates"] = (
    state_month["total_demo_updates"] +
    state_month["total_bio_updates"]
)
```

```python
state_month["update_to_enrolment_ratio"] = (
    state_month["total_updates"] /
    state_month["total_enrolments"].replace(0, pd.NA)
)
```

```python
import matplotlib.pyplot as plt

national_trend = (
    state_month
    .groupby("year_month")[["total_updates", "total_enrolments"]]
    .sum()
    .reset_index()
)

national_trend["ratio"] = (
    national_trend["total_updates"] /
    national_trend["total_enrolments"]
)

plt.figure(figsize=(10,5))
plt.plot(
    national_trend["year_month"].astype(str),
    national_trend["ratio"]
)
plt.axhline(1, linestyle="--")

plt.xticks(rotation=45)
plt.ylabel("Update-to-Enrolment Ratio")
plt.title("Is Aadhaar a One-Time ID or a Maintenance System? (National Trend)")
plt.tight_layout()
plt.show()
```

```python
state_avg = (
    state_month
    .groupby("state")["update_to_enrolment_ratio"]
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(8,5))
state_avg.plot(kind="bar")
plt.axhline(1, linestyle="--")

plt.ylabel("Avg Update-to-Enrolment Ratio")
plt.title("States Where Aadhaar Acts as a Maintenance System")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
focus_states = ["Bihar", "Rajasthan", "Tamil Nadu", "Assam", "Maharashtra"]

subset = state_month[state_month["state"].isin(focus_states)]

plt.figure(figsize=(10,5))
for state in focus_states:
    s = subset[subset["state"] == state]
    plt.plot(s["year_month"].astype(str), s["update_to_enrolment_ratio"], label=state)

plt.axhline(1, linestyle="--")
plt.legend()
plt.xticks(rotation=45)
plt.ylabel("Update-to-Enrolment Ratio")
plt.title("Aadhaar Maintenance Burden Across States")
plt.tight_layout()
plt.show()
```

```python
bio_dm = (
    bio.groupby(["state", "district", "year_month"])["total_bio_updates"]
    .sum()
    .reset_index()
)
```

```python
enrol_dm = (
    enrol.groupby(["state", "district", "year_month"])["total_enrolments"]
    .sum()
    .reset_index()
)
```

```python
bio_dm = bio_dm.merge(
    enrol_dm,
    on=["state", "district", "year_month"],
    how="left"
)

bio_dm["biometric_updates_per_1000"] = (
    bio_dm["total_bio_updates"] /
    bio_dm["total_enrolments"].replace(0, pd.NA)
) * 1000
```

```python
import matplotlib.pyplot as plt

national_bio = (
    bio_dm
    .groupby("year_month")["biometric_updates_per_1000"]
    .mean()
    .reset_index()
)

plt.figure(figsize=(9,5))
plt.plot(
    national_bio["year_month"].astype(str),
    national_bio["biometric_updates_per_1000"]
)

plt.xticks(rotation=45)
plt.ylabel("Biometric Updates per 1,000 Enrolled Residents")
plt.title("National Trend in Biometric Update Intensity")
plt.tight_layout()
plt.show()
```

```python
cutoff = bio_dm["year_month"].sort_values().unique()[2]  # early 23 months

early = bio_dm[bio_dm["year_month"] <= cutoff]
late  = bio_dm[bio_dm["year_month"] > cutoff]
```

```python
early_avg = (
    early.groupby(["state", "district"])["biometric_updates_per_1000"]
    .mean()
    .reset_index(name="early_bio")
)

late_avg = (
    late.groupby(["state", "district"])["biometric_updates_per_1000"]
    .mean()
    .reset_index(name="late_bio")
)

bio_persistence = early_avg.merge(
    late_avg, on=["state", "district"]
)
```

```python
plt.figure(figsize=(7,6))
plt.scatter(
    bio_persistence["early_bio"],
    bio_persistence["late_bio"],
    alpha=0.5
)

plt.xlabel("Early-Period Biometric Updates per 1,000")
plt.ylabel("Late-Period Biometric Updates per 1,000")
plt.title("Biometric Stability Test: Early vs Late Period")
plt.tight_layout()
plt.show()
```

```python
bio_persistence["persistence_ratio"] = (
    bio_persistence["late_bio"] /
    bio_persistence["early_bio"].replace(0, pd.NA)
)

high_persistence = (
    bio_persistence
    .sort_values("persistence_ratio", ascending=False)
    .head(15)
)

plt.figure(figsize=(10,6))
plt.barh(
    high_persistence["district"] + ", " + high_persistence["state"],
    high_persistence["persistence_ratio"]
)

plt.xlabel("Late / Early Biometric Update Ratio")
plt.title("Districts Where Biometric Updates Remain Structurally High")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import spearmanr

valid = bio_persistence.dropna()

corr, pval = spearmanr(
    valid["early_bio"],
    valid["late_bio"]
)

print("Spearman correlation:", corr)
print("p-value:", pval)
```
